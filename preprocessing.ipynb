```python
import pandas as pd 
```

```python
pre_course = pd.read_excel("Pre-course survey-NoName.xlsx", skiprows=1)
```

```python
# Preprocessing
print(f"Pre-course survey has duplicated entries? {pre_course['Student Number'].nunique() != len(pre_course)}")
# no duplicates 

for column in pre_course.columns:
    if 'Q2' in column:
        pre_course[column].fillna('No answer', inplace=True)

stu_no_enumerated = list(enumerate(pre_course['Student Number'].values, start = 1))
stu_dict = {}
for tup in stu_no_enumerated:
    stu_dict[tup[1]] = tup[0]

pre_course['Student Number'] = pre_course['Student Number'].map(stu_dict)

q3_columns = [col for col in pre_course.columns if 'Q3' in col]
q4_columns = [col for col in pre_course.columns if 'Q4' in col]

q3_columns_enum = list(enumerate(q3_columns, start = 1))
q4_columns_enum = list(enumerate(q4_columns, start = 1))


q3_dict = {}
for tup in q3_columns_enum:
    q3_dict[tup[1]] = f"Q3.{tup[0]}"

q4_dict = {}
for tup in q4_columns_enum:
    q4_dict[tup[1]] = f"Q4.{tup[0]}"

pre_course.rename(columns = q3_dict, inplace = True)
pre_course.rename(columns = q4_dict, inplace = True)

pre_course.head()

```

```python
post_course = pd.read_excel("Post-course survey-NoName.xlsx", skiprows=1)
```
